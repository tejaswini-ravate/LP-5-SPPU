```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
```

```python
# 1. Load CSV data
train_df = pd.read_csv("mnist_fashion_train.csv")
test_df  = pd.read_csv("mnist_fashion_test.csv")

print(train_df)
print(test_df)
```

```python
# 2. Split into features (X) and labels (y)
x_train = train_df.iloc[:, 1:].values   # pixels
y_train = train_df.iloc[:, 0].values    # labels 0-9

x_test  = test_df.iloc[:, 1:].values
y_test  = test_df.iloc[:, 0].values
```

```python
# 3. Reshape and normalize
x_train = x_train.reshape((60000, 28, 28, 1)).astype("float32") / 255.0
x_test = x_test.reshape((10000, 28, 28, 1)).astype("float32") / 255.0
```

```python
# 4. One-hot encode labels
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)
```

```python
# 5. Class names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# 6. Simple CNN model
model = Sequential()

model.add(Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))   # 10 classes

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# 7. Train model
history = model.fit(
    x_train, y_train_cat,
    epochs=5,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)
```

```python
# 8. Evaluate on test data
test_loss, test_acc = model.evaluate(x_test, y_test_cat, verbose=0)
print("\nTest Loss:", test_loss)
print("Test Accuracy:", test_acc)
```
